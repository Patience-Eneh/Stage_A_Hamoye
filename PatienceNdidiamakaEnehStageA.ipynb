```python
import pandas as pd

#Question 2

# Load the CSV file into a pandas DataFrame
df = pd.read_csv("/Users/user/Downloads/FoodBalanceSheets_E_Africa_NOFLAG.csv", encoding='latin-1')

# Extract the year columns and 'Element Code' column
year_columns = ['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']
element_code_column = 'Element Code'

# Calculate the correlation between 'Element Code' and other numerical columns for each year
correlation_by_year = {}
for year_col in year_columns:
    correlation = df[[year_col, element_code_column]].corr().iloc[0, 1]
    correlation_by_year[year_col] = correlation

# I will find the year with the least correlation
least_correlated_year = min(correlation_by_year, key=correlation_by_year.get)


print(least_correlated_year)


#QUESTION 4

#number of missig values in the 'Y2014' column
missing_values_count = df['Y2014'].isnull().sum()

#total number of entries in the 'Y2014' column
total_entries = df['Y2014'].shape[0]

# the percentage of missing values
missing_percentage = (missing_values_count / total_entries) * 100


print(missing_values_count)
print("Percentage of missing values in Y2014: {:.3f}%".format(missing_percentage))


#QUESTION 7


# Count the number of unique areas (countries)
total_unique_areas = df['Area'].nunique()


print(total_unique_areas)


#QUESTION 9

# Select columns 'Y2017' and 'Area'
selected_columns = ['Y2017', 'Area']
df_selected = df[selected_columns]

# Perform groupby operation on 'Area' and calculate sum for 'Y2017'
sum_of_area = df_selected.groupby('Area')['Y2017'].sum()

# Find the area with the highest sum in 2017
area_with_highest_sum = sum_of_area.idxmax()

print("Area with the highest sum in 2017:", area_with_highest_sum)


#QUESTION 10


# Perform groupby operation on 'Item' and calculate sum for each year
sum_by_itemyear = df.pivot_table(index='Item', values=['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018'], aggfunc='sum')

#the year with the highest sum of Stock Variation
year_with_highest_sum = sum_by_itemyear.sum(axis=0).idxmax()

print("Year with the highest sum of Stock Variation:", year_with_highest_sum)


#QUESTION 11


# Calculate the total sum of Wine produced in Y2015
wine_2015_sum = df['Y2015'][df['Item'] == 'Wine'].sum()

# Calculate the total sum of Wine produced in Y2018
wine_2018_sum = df['Y2018'][df['Item'] == 'Wine'].sum()

print("Total sum of Wine produced in Y2015:", wine_2015_sum)
print("Total sum of Wine produced in Y2018:", wine_2018_sum)



#QUESTION 14

# Filter the DataFrame for rows where the 'Element' column is 'Processing' and the year is 2017
processing_2017 = df[df['Element'] == 'Processing'].groupby('Element')['Y2017'].sum()

# Display the result
print("Total sum of Processing in 2017:", processing_2017)

#QUESTION 16

# Calculate the mean and standard deviation for the 'Y2017' column
mean_2017 = df['Y2017'].mean()
std_dev_2017 = df['Y2017'].std()

# Round the values to 2 decimal places
mean_2017 = round(mean_2017, 2)
std_dev_2017 = round(std_dev_2017, 2)

print("Mean for the year 2017:", mean_2017)
print("Standard deviation for the year 2017:", std_dev_2017)



#QUESTION 17

# Select columns 'Y2017' and 'Area'
selected_columns = ['Y2017', 'Area']
df_selected = df[selected_columns]

# Perform groupby operation on 'Area' and calculate sum for 'Y2017'
sum_by_area_2017 = df_selected.groupby('Area')['Y2017'].sum()

# Sort the sums in ascending order and select the 7th lowest sum
seventh_lowest_sum = sum_by_area_2017.nsmallest(7).iloc[-1]

# Find the area with the 7th lowest sum in 2017
area_with_seventh_lowest_sum = sum_by_area_2017[sum_by_area_2017 == seventh_lowest_sum].index[0]


print("Area with the 7th lowest sum in 2017:", area_with_seventh_lowest_sum)





```
